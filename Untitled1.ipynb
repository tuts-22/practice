```python
import pandas as pd
import numpy as np
```

```python
orig_data = pd.read_csv("features_data.csv")

print(orig_data.shape)
orig_data.head()
```

```python
cols = orig_data.columns
print(cols[:3])
weeks = orig_data[3:]

new_df = pd.melt(orig_data, id_vars=cols[:3], var_name="Date", value_name="Value")

new_df["Measures"] = new_df["Causal"] + " " + new_df["Measures"]
new_df.drop("Causal", axis=1, inplace=True)
new_df.head(-100)
```

```python
temp = pd.concat([new_df, pd.DataFrame()])
temp = temp.sort_values('Date')

temp = temp[temp["Measures"] != "Causal Measures"]
temp.head()

for i, row in temp.iterrows():
    date = row["Date"]
    date = date.split(" ")[-1]
    if date.endswith("16.1"):
        date = date.replace("16.1", "15")
    if date.endswith("16.2"):
        date = date.replace("16.2", "14")
    if date.endswith("16.3"):
        date = date.replace("16.3", "13")
    temp.set_value(i, "Date", date)

temp.head(-100)
```

```python
temp["Value"] = temp["Value"].astype(float)
t = temp.pivot_table('Value', ["Product", "Date"], 'Measures')
t = t.reset_index()
t = t.drop_duplicates()
```

```python
print(t.shape)
t.head(100)
```

```python
rest = pd.read_csv("merged_ext_with_product.csv")
rest.shape
```

```python
merged = pd.merge(rest, t, how="left", left_on=["Product", "Date"], right_on=["Product", "Date"])
merged.shape
```

```python
merged.to_csv("final_merged.csv")
```
